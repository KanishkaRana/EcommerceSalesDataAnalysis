```python
data = pd.read_csv('data.csv', encoding='ISO-8859-1')
```

```python
data.head()
```

```python
query = """
Select * from data
where InvoiceNo is null
"""
nullinvoice = ps.sqldf(query, locals())
nullinvoice
#no null values in invoice column
```

```python
data['UnitPriceFormatted'] = data['UnitPrice'].apply(lambda x: "{:.2f}$".format(x))
data.head()
```

```python
data.drop(columns=['UnitPrice'], inplace=True)
data.head()
#dropped the unit price column because we have unitpriceformatted.
```

```python
q2= """
select sum(Quantity * UnitPriceFormatted) as TotalRevenue
from data
"""
TotalRevenue =ps.sqldf(q2, locals())
TotalRevenue
#total revenue calculated
```

```python
#top 10 customers
q3 = """
select CustomerID, sum(Quantity * UnitPriceFormatted) as TotalRevenue
from data
group by CustomerID
order by TotalRevenue desc
limit 10
"""
top10customers=ps.sqldf(q3, locals())
top10customers
```

```python
#top 10 selling products
q4 = """
select Description, sum(Quantity * UnitPriceFormatted) as TotalRevenue
from data
where Description is not null
group by Description
order by TotalRevenue desc
limit 10
"""
top10products = ps.sqldf(q4, locals())
top10products
```
